```python
import pyesg
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
s1 = time.time()
model1 = pyesg.CoxIngersollRossProcess(mu=0.0546, sigma=0.0412,theta=1.0682)
results1 = model1.scenarios(0.0228, 1/12, 10000, 1200, 1)

model2 = pyesg.CoxIngersollRossProcess(mu=0.0778, sigma=0.0287,theta=0.0469)
results2 = model2.scenarios(0.0809, 1/12, 10000, 1200, 10)

phi = 0.45 - 0.0809 -  0.0228
n = phi + results1 + results2
e1 = time.time()
```

```python
print(e1-s1)
```

```python
for i in range(1000):
    plt.plot(n[i,:],linewidth=0.5)
plt.xlabel("Time")
plt.ylabel("Short Rate")
```
